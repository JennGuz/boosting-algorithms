```python
# Your code here
import pandas as pd 
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from pickle import dump

diabetes_test_data = pd.read_csv("https://raw.githubusercontent.com/JennGuz/decision-tree/refs/heads/main/data/processed/clean_test.csv")
diabetes_train_data = pd.read_csv("https://raw.githubusercontent.com/JennGuz/decision-tree/refs/heads/main/data/processed/clean_train.csv")
diabetes_train_data.sample(10)
```

```python
diabetes_test_data.sample(10)
```

```python
X_train = diabetes_train_data.drop(columns="Outcome")
y_train = diabetes_train_data["Outcome"]
X_test = diabetes_train_data.drop(columns="Outcome")
y_test = diabetes_train_data["Outcome"]
```

```python
model = XGBClassifier(random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
dump(model, open("boosting_algortithms_42.sav", "wb"))
```
